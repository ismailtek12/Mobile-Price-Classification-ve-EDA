#### Outlier Detection

```python
def Boxplot(df,col):
    plt.figure(figsize=(6,6))
    sns.boxplot(x=col,data=df)
    plt.show()
```

```python
for i in num_list:
    Boxplot(df,i)
```

Ã–n kamera kalitesi ve pixel Ã§Ã¶zÃ¼nÃ¼rlÃ¼k yÃ¼kseliÄŸi iÃ§erisinde bazÄ± outlier deÄŸerler var ama bunlarÄ± Ã§Ä±kartmak ne kadar doÄŸru modelleyerk Ã¶ÄŸrenebiliriz.

```python
import numpy as np
def outlier(df,col):
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    alt=q1-1.5*iqr
    ust=q3+1.5*iqr
    out=((df[col]<alt)|(df[col]>ust))
    index=df[out].index
    #value=df[out].values
    return index
```

```python
a=outlier(df,"fc")
type(a)
```

```python
#OutlierlarÄ± temizlemeden Ã¶nce karÅŸÄ±laÅŸtÄ±rmak iÃ§in veriyi kopyalayalÄ±m
df_ex_out=df.copy()
```

```python
df_ex_out["fc"][a]=df_ex_out["fc"].mean()
```
